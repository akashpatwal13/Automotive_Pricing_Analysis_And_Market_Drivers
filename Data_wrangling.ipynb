```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Import scaling tools for later steps
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define column names (since dataset has no header row)
cols = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration", 
        "num_doors", "body_style", "drive_wheels", "engine_location", "wheel_base", 
        "length", "width", "height", "curb_weight", "engine_type", "num_cylinders", 
        "engine_size", "fuel_system", "bore", "stroke", "compression_ratio", 
        "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]

# Load dataset from UCI repository
# Replace '?' values with NaN so pandas can recognize missing values
df = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",
    names=cols,
    na_values="?"
)

# Display first few rows
df.head()
```

```python
# Count missing values in each column
df.isna().sum()
```

```python
# Get dataset info: data types and non-null counts
df.info()
```

```python
# Remove rows where 'price' or 'horsepower' is missing
# These are important variables for analysis/modeling
df = df.dropna(subset=["price", "horsepower"])
```

```python
# List of numeric columns with missing values
numeric_cols = ["normalized_losses", "bore", "stroke", "horsepower", "peak_rpm"]

# Replace missing values with median of each column
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# List of categorical columns with missing values
categorical_cols = ["num_doors", "fuel_system"]

# Replace missing values with most frequent value (mode)
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Convert selected columns to numeric datatype
cols_to_numeric = ["normalized_losses", "bore", "stroke", "horsepower", "peak_rpm", "price"]

for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col])
```

```python
# Initialize StandardScaler
scaler = StandardScaler()

# Select numeric columns
num_features = df.select_dtypes(include=["float64", "int64"]).columns

# Create a copy of dataset for standardized values
df_standardized = df.copy()

# Apply Z-score standardization
df_standardized[num_features] = scaler.fit_transform(df[num_features])

# Preview standardized data
df_standardized.head()
```

```python
# Initialize MinMaxScaler
minmax = MinMaxScaler()

# Create a copy of dataset for normalized values
df_normalized = df.copy()

# Scale numeric values between 0 and 1
df_normalized[num_features] = minmax.fit_transform(df[num_features])

# Preview normalized data
df_normalized.head()
```

```python
# Create a copy for binning
df_binned = df.copy()

# Divide horsepower into 3 categories: Low, Medium, High
df_binned["hp_bin"] = pd.cut(
    df_binned["horsepower"],
    bins=3,
    labels=["Low", "Medium", "High"]
)

# View result
df_binned[["horsepower", "hp_bin"]].head(10)
```

```python
# Convert categorical columns into dummy/indicator variables
df_encoded = pd.get_dummies(
    df,
    columns=["fuel_type", "aspiration", "body_style"],
    drop_first=True  # Avoid dummy variable trap
)

# Preview encoded dataset
df_encoded.head()
```

```python
# Save final cleaned and processed dataset to CSV file
df_encoded.to_csv("cleaned_auto_data_final.csv", index=False)
```
